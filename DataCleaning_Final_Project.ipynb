```python
#subsetting the data based on date selected for each month and delta
def subsetting_data(df,fromdate, todate, frequency, delta):
  dates = pd.date_range(start= fromdate, end=todate, freq=frequency)
  date_rows = df[df.index.isin(dates)]
  date_rows = date_rows[abs(date_rows['Delta of the Option']) == delta]
  return date_rows
```

```python
#getting first friday of everymonth for entire dataset
fromdate = '2005-01-01'
todate = '2023-01-31'
delta = 50
```

## Data Sanity Check

```python
# Filter rows to make sure the high is the high and low is the low
subset_data = subset_data[subset_data['Low'] < subset_data['High']]
subset_data = subset_data[subset_data['High'] > subset_data['Low']]
subset_data = subset_data[subset_data['Low'] <= subset_data['Open']]
subset_data = subset_data[subset_data['Low'] <= subset_data['Close']]
subset_data = subset_data[subset_data['High'] >= subset_data['Open']]
subset_data = subset_data[subset_data['High'] >= subset_data['Close']]
len(subset_data)
```

## Calculating Implied Volatilities

```python
def calc_implied_vols(df):
  implied_vols = []
  unique_index_values = df.index.unique()
  for date in unique_index_values:
    subset = df.loc[[date]]
    implied_vols.append(subset['Interpolated Implied Volatility of the Option'].mean())
  implied_vol_df = pd.DataFrame({'Implied Vols': implied_vols}, index=unique_index_values)
  return implied_vol_df
```

```python
implied_voldf = calc_implied_vols(subset_data)
```

## Calculating Historical Volatilities 6 Month

```python
historical_vol_data = stock_data
```

```python
historical_vol_data['Date'] = pd.to_datetime(historical_vol_data['Date'])
historical_vol_data.set_index('Date', inplace=True)
```

```python
#calculating log returns
returns_data = np.log(historical_vol_data['Adj Close']/historical_vol_data['Adj Close'].shift(1))
```

```python
#getting rid of first nan value
returns_data = returns_data.tail(-1)
```

```python
returns_data
```

```python
#rolling standard deviation
def rolling_std(returns_df, time_frame, startdate):
  date_list = []
  std_list = []
  #mean_list = []
  for i in range(0, 6200):
    to_date = pd.to_datetime(startdate)
    to_date = to_date + relativedelta(days=i)
    from_date = to_date - relativedelta(months=time_frame)
    date_list.append(to_date)
    selected_df = returns_df[returns_df.index.to_series().between(from_date, to_date)]
    #selected_df = returns_df.query("Date >= @from_date & Date <= @to_date").copy()
    std_value = selected_df.std() * np.sqrt(252)
    #mean_value = selected_df.mean() * 252
    std_list.append(std_value)
    #mean_list.append(mean_value)
  #rolling_std_df = pd.DataFrame({'Historical Volatility': std_list, 'Historical Mean' : mean_list}, index=date_list)
  rolling_std_df = pd.DataFrame({'Historical Volatility': std_list}, index=date_list)
  return rolling_std_df
```

```python
# 6 month historical standard deviation
historical_data_sixmonth = rolling_std(returns_data, 6, '2006-01-01')
```

```python
historical_data_sixmonth.rename(columns={'index': 'Date'}, inplace=True)
```

```python
#merge historical data with implied volatility data
final_merged_data_sixmonth = pd.merge(historical_data_sixmonth, implied_voldf, on = 'Date', how = 'inner')
```

```python
#annualize risk free rate
final_merged_data_sixmonth['RF'] = final_merged_data_sixmonth['RF'] * 252
```

```python
#expected mean calculated based on CAPM; Beta = 1; Equity Risk Premium of 5.75%
final_merged_data_sixmonth['Expected Mean'] = final_merged_data_sixmonth['RF'] + 1 * .0575
```

```python
#merge historical data with implied volatility data
final_merged_data_sixmonth = pd.merge(final_merged_data_sixmonth, historical_vol_data, on = 'Date', how = 'inner')
```
